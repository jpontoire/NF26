## Object Creation



```python
from datetime import timedelta
from dateutil import parser

def generation_data_patients(n): 

    # Table Administration
    key_consultation = [] #OK
    key_patient = [] #OK
    name_patients = [] #OK
    firstname_patients = [] #OK
    Num_Secu = [] # OK
    dates_In = [] #OK
    pathology = [] #OK
    dates_Out = [] #OK
    
    # Table Medecin
    nbMedecin = 30 #OK
    key_medecin = [] #OK
    name_medecin = [] #OK
    firstname_medecin = [] #OK
    speciality = ['Generaliste','Chirurgien','Dermatologiste','Dentiste','Cardiologue'] #OK
    specialyty_Medecin = [] #OK
    for l in range(nbMedecin): 
        key_medecin.append('keyMedecin' + str(l)) #OK
        name_medecin.append('NameMedecin' + str(l)) #OK
        firstname_medecin.append('FirstNameMedecin' + str(l)) #OK
        idxSpecility = np.random.poisson(lam=1, size=1)[0] #OK
        if idxSpecility > 4:
            idxSpecility = 4 #OK
        specialyty_Medecin.append(speciality[idxSpecility]) #OK
    MedecinConsult = [] #OK
    
    # Patient Profile
    age = [] #OK
    weight = [] #OK
    temperature = [] #OK
    tension = [] #OK
    diabete = [] #OK
    
    # Table Treatment
    key_treatment_Patient = [] #OK
    key_treatment = [] #OK
    treatment = [] #OK
    key_medicament = [] #OK
    name_medicament = [] #OK
    quantity_medicament = [] #OK
    for t in range(1000):
        key_medicament.append('KeyMed' + str(t))
        name_medicament.append('Medicament' + str(t))
    
    # Table Chambre
    key_chambre = [] #OK
    chambre_number = [] #OK
    numberChambre = 104
    for c in range(numberChambre):
        key_chambre.append('KeyChambre' + str(c+1))
        chambre_number.append('C' + str(f'{c+1:03}'))
    chambre_Number_Patient = [] #OK
    key_chambre_Patient = [] #OK
    
    
    
    year = 2023
    str_year = str(year)
    all_dates_in = np.sort(np.random.randint(1, 365, n))
    
    for i in range(n):
        
        # key Consultation :
        key_consultation.append('KeyConsult' + str(i))
        
        # key Patient :
        idx_patient = str(np.random.randint(1, 50000, 1)[0])
        key_patient.append('KeyPatient' + idx_patient)
        
        # Num Secu :
        SecuOK = np.random.binomial(1, 0.99, 1)[0]
        if SecuOK == 1:
            Num_Secu.append('NS' + str(f'{i+1:05}'))
        if SecuOK == 0:
            Num_Secu.append('NULL')
    
        # Name :
        name_patients.append('Name' + idx_patient)

        # First Name :
        firstname_patients.append('FistName' + idx_patient)
    
        # Pathology :
        idx_pathology = np.random.randint(1, 100, 1)[0]
        pathology.append('Pathology' + str(idx_pathology))
        
        # Medecin :
        MedecinConsult.append(key_medecin[np.random.randint(1, nbMedecin, 1)[0]])
        
        # initializing year and day number
        day_num_In = all_dates_in[i]
        
        if idx_pathology <= 60 :
            lmbda = 1.0
            # Age :
            age.append(np.random.poisson(lam=20, size=1)[0])
            # Weight :
            weight.append(np.random.normal(60, 3, 1)[0])
            # Temperature :
            temperature.append(np.random.normal(37.5, 0.1, 1)[0])
            # Tension :
            tension.append(np.random.normal(116, 20, 1)[0])
            # Diabete :
            diabete.append(np.random.binomial(1, 0.1, 1)[0])
            # Treatment :
            nbMed = np.random.randint(1, 4, 1)[0]
            idxMed = np.random.randint(1, 1000, nbMed)
            for m in range(nbMed):
                key_treatment.append('KeyTreatment' + str(i))
                treatment.append(key_medicament[idxMed[m]])
                quantity_medicament.append(np.random.randint(1, 4, 1)[0])
        if idx_pathology > 60 and idx_pathology <= 70 :
            lmbda = 3.0
            # Age :
            age.append(np.random.poisson(lam=40, size=1)[0])
            # Weight :
            weight.append(np.random.normal(75, 7, 1)[0])
            # Temperature :
            temperature.append(np.random.normal(38.5, 0.1, 1)[0])
            # Tension :
            tension.append(np.random.normal(110, 5, 1)[0])
            # Diabete :
            diabete.append(np.random.binomial(1, 0.1, 1)[0])
            # Treatment :
            nbMed = np.random.randint(1, 4, 1)[0]
            idxMed = np.random.randint(1, 5, nbMed)
            for m in range(nbMed):
                key_treatment.append('KeyTreatment' + str(i))
                treatment.append(key_medicament[idxMed[m]])
                quantity_medicament.append(np.random.randint(1, 4, 1)[0])
        if idx_pathology > 70 and idx_pathology <= 80 :
            lmbda = 6.0
            # Age :
            age.append(np.random.poisson(lam=70, size=1)[0])
            # Weight :
            weight.append(np.random.normal(70, 4, 1)[0])
            # Temperature :
            temperature.append(np.random.normal(40, 0.1, 1)[0])
            # Tension :
            tension.append(np.random.normal(120, 5, 1)[0])
            # Diabete :
            diabete.append(np.random.binomial(1, 0.2, 1)[0])
            # Treatment :
            nbMed = np.random.randint(1, 4, 1)[0]
            idxMed = np.random.randint(1, 5, nbMed)
            for m in range(nbMed):
                key_treatment.append('KeyTreatment' + str(i))
                treatment.append(key_medicament[idxMed[m]])
                quantity_medicament.append(np.random.randint(1, 4, 1)[0])
        if idx_pathology > 80 and idx_pathology <= 90 :
            lmbda = 10.0
            # Age :
            age.append(np.random.poisson(lam=50, size=1)[0])
            # Weight :
            weight.append(np.random.normal(80, 3, 1)[0])
            # Temperature :
            temperature.append(np.random.normal(38, 0.1, 1)[0])
            # Tension :
            tension.append(np.random.normal(90, 2, 1)[0])
            # Diabete :
            diabete.append(np.random.binomial(1, 0.7, 1)[0])
            # Treatment :
            nbMed = np.random.randint(1, 4, 1)[0]
            idxMed = np.random.randint(1, 90, nbMed)
            for m in range(nbMed):
                key_treatment.append('KeyTreatment' + str(i))
                treatment.append(key_medicament[idxMed[m]])
                quantity_medicament.append(np.random.randint(1, 4, 1)[0])
        if idx_pathology > 90 and idx_pathology <= 100 :
            lmbda = 18.0
            # Age :
            age.append(np.random.poisson(lam=40, size=1)[0])
            # Weight :
            weight.append(np.random.normal(75, 4, 1)[0])
            # Temperature :
            temperature.append(np.random.normal(40.5, 0.1, 1)[0])
            # Tension :
            tension.append(np.random.normal(80, 2, 1)[0])
            # Diabete :
            diabete.append(np.random.binomial(1, 0.2, 1)[0])
            # Treatment :
            nbMed = np.random.randint(1, 4, 1)[0]
            idxMed = np.random.randint(1, 1000, nbMed)
            for m in range(nbMed):
                key_treatment.append('KeyTreatment' + str(i))
                treatment.append(key_medicament[idxMed[m]])
                quantity_medicament.append(np.random.randint(1, 4, 1)[0])
                
        key_treatment_Patient.append('KeyTreatment' + str(i))
            
        number_days_stay = np.random.poisson(lam=lmbda, size=1)[0]
        day_num_Out = day_num_In + number_days_stay
            

        # creating date string
        date_str = str_year + "-01-01"  # January 1st of the given year
        date_obj = parser.parse(date_str)  # parse date string to datetime object

        # adding days to datetime object for date_In
        date_res_In = date_obj + timedelta(days=int(day_num_In) - 1)
        dates_In.append(date_res_In)
        
        # adding days to datetime object for date_Out
        date_res_out = date_obj + timedelta(days=int(day_num_Out) - 1)
        dates_Out.append(date_res_out)


        # Chambre ?
        if number_days_stay > 0:
            idxChambre = np.random.randint(numberChambre, size=1)[0]
            key_chambre_Patient.append(key_chambre[idxChambre])
            chambre_Number_Patient.append(chambre_number[idxChambre])
        if number_days_stay == 0:
            key_chambre_Patient.append('NULL')
            chambre_Number_Patient.append('NULL')
        
        
        
        
    # Table Administration
    pandas_administration_df = pd.DataFrame(
        {'KeyConsult': key_consultation,
         'KeyPatient': key_patient,
         'Name': name_patients,
         'FirstName': firstname_patients,
         'NumSecu': Num_Secu,
         'Date_In': dates_In,
         'Pathology': pathology,
         'KeyChambre': key_chambre_Patient,
         'ChambreNumber': chambre_Number_Patient,
         'Date_Out': dates_Out
        }
    )


    # Table Medecin
    pandas_medecin_df = pd.DataFrame(
        {'KeyMedecin': key_medecin,
         'NameMedecin': name_medecin,
         'FirstNameMedecin': firstname_medecin,
         'SpecialityMedecin': specialyty_Medecin
        }
    )


    # Table Diagnostic
    pandas_diagnostic_df = pd.DataFrame(
        {'KeyConsult': key_consultation,
         'KeyMedecin': MedecinConsult,
         'KeyPatient': key_patient,
         'NamePatient': name_patients,
         'FirstNamePatient': firstname_patients,
         'NumSecu': Num_Secu,
         'Age': age,
         'Weight': weight,
         'Temperature': temperature,
         'Tension': tension,
         'Diabete': diabete,
         'Pathology': pathology,
         'KeyTreatment': key_treatment_Patient
        }
    )


    # Table Treatment
    pandas_treatment_df = pd.DataFrame(
        {'KeyTreatment': key_treatment,
         'KeyMedicament': treatment,
         'QuantityMedicament': quantity_medicament
        }
    )
    
    
    # Table Medicaments
    pandas_medicament_df = pd.DataFrame(
        {'KeyMedicament': key_medicament,
         'NameMedicament': name_medicament
        }
    )


    # Table Chambre
    pandas_chambre_df = pd.DataFrame(
        {'KeyChambre': key_chambre,
         'ChambreNumber': chambre_number
        }
    )

        
        
        
    return pandas_administration_df, pandas_medecin_df, pandas_diagnostic_df, pandas_treatment_df, pandas_medicament_df, pandas_chambre_df






```

```python
pandas_administration_df, pandas_medecin_df, pandas_diagnostic_df, pandas_treatment_df, pandas_medicament_df, pandas_chambre_df = generation_data_patients(50000)
```
